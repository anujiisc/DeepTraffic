##### Convert Dataset Annotations in MOT challenge format

```python
import os
import xml.etree.ElementTree as ET

def convert_annotations(xml_folder, output_file):
    with open(output_file, 'w') as f_out:
        
        li = os.listdir(xml_folder)
        li = [int(e[:-4]) for e in li]
        li.sort()
        li = [str(e)+'.xml' for e in li][:1000]
        frame = 1
        for filename in li:
            if filename.endswith('.xml'):
                tree = ET.parse(os.path.join(xml_folder, filename))
                root = tree.getroot()

                for obj in root.findall('object'):
                    class_name = obj.find('class').text
                    class_id = class_mapping.get(class_name, -1)
                    if class_id == -1:
                        continue
                    
                    xmin = int(obj.find('bndbox/xmin').text)
                    ymin = int(obj.find('bndbox/ymin').text)
                    xmax = int(obj.find('bndbox/xmax').text)
                    ymax = int(obj.find('bndbox/ymax').text)

                    # Convert to MOT16 format
                    
                    obj_id = int(obj.find('ID').text)
                    bb_left = xmin
                    bb_top = ymin
                    bb_width = xmax - xmin
                    bb_height = ymax - ymin
                    confidence = 1  # Assuming confidence is always 1
                    x, y, z = -1, -1, -1  # Set world coordinates to -1

                    # Write the MOT16 formatted annotation to output file
                    f_out.write(f"{frame},{obj_id},{bb_left},{bb_top},{bb_width},{bb_height},{confidence},{x},{y},{z}\n")
                frame += 1

# Define class mapping (replace with your own)
class_mapping = {'car': 0, 'motorbike': 1, 'van': 2, 'truck': 3, 'big-truck': 4, 'bus': 5}

# Specify the folder containing XML annotations and the output file
xml_folder = '../Data/GRAM/GRAM-RTMv4/Annotations/M-30/xml/'
output_file = './GRAM_MOT1/M-30.txt'

# Convert annotations
convert_annotations(xml_folder, output_file)

# Specify the folder containing XML annotations and the output file
xml_folder = '../Data/GRAM/GRAM-RTMv4/Annotations/M-30-HD/xml/'
output_file = './GRAM_MOT1/M-30-HD.txt'

# Convert annotations
convert_annotations(xml_folder, output_file)


# Specify the folder containing XML annotations and the output file
xml_folder = '../Data/GRAM/GRAM-RTMv4/Annotations/Urban1/xml/'
output_file = './GRAM_MOT1/Urban1.txt'

# Convert annotations
convert_annotations(xml_folder, output_file)
```
